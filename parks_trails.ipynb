```python
import pandas as pd
df = pd.read_csv('parks_trails.csv')
```

```python
df.head()
```

```python
df.iff_reclass.unique()
# there are nan(s) in this column
```

```python
df.fillna('Unlisted', inplace=True)
df.head()
```

```python
df.rename(columns={'MILEAGE_ROUNDED': 'Total Miles', 'PARK_NAME': 'Park Name', 'iff_reclass': 'Difficulty', 'TRAIL_NAME': 'Trail Name'}, inplace=True)
df.head()
```

```python
df.Difficulty.unique()
```

```python
df['Trail Name'].unique
```

```python
import numpy as np
table = pd.pivot_table(data=df, values='Total Miles', index=['Park Name', 'Difficulty'], aggfunc=np.sum)
table
```

```python
table2 = pd.pivot_table(data=df, values='Trail Name', index=['Park Name', 'Difficulty'], aggfunc=lambda x: ', '.join(x))
table2
```

```python
#drop rows where 'Unlisted' is the only value
df_no_unl = df[df['Trail Name'] != 'Unlisted']
df_no_unl
```

```python
table_3 = pd.pivot_table(data=df_no_unl, values='Trail Name', index=['Park Name', 'Difficulty'], aggfunc=lambda x: ', '.join(x))
table_3
```

```python
df_no_dupes = df_no_unl.drop_duplicates(subset='Trail Name')
df_no_dupes
```

```python
table_4 = pd.pivot_table(data=df_no_unl, values='Trail Name', index=['Park Name', 'Difficulty'], aggfunc=lambda x: ', '.join(list(set(x))))
table_4
```

```python
table_4.to_excel('final_output.xlsx')
```

```python
df_final = pd.DataFrame(table_4)
df_final.head()
```
