```python
import pandas as pd
df = pd.read_csv('parks_trails.csv')
df.head()
```

```python
df.iff_reclass.unique()
# NaN in this column - gotta replace
```

```python
df.fillna('Unlisted', inplace=True)
df.head()
```

```python
df.rename(columns={'MILEAGE_ROUNDED': 'Total Miles', 'PARK_NAME': 'Park Name', 'iff_reclass': 'Difficulty', 'TRAIL_NAME': 'Trail Name'}, inplace=True)
df.head()
```

```python
df.Difficulty.unique()
```

```python
df['Trail Name'].unique
```

```python
#drop rows where 'Unlisted' is the only value
df = df[df['Trail Name'] != 'Unlisted']
df.head()
```

```python
# get total mileage per taril rating per park
import numpy as np
dist_by_diff_table = pd.pivot_table(data=df, values='Total Miles', index=['Park Name', 'Difficulty'], aggfunc=np.sum)
dist_by_diff_table
```

```python
# get trail names by difficulty per park
names_by_diff_table = pd.pivot_table(data=df, values='Trail Name', index=['Park Name', 'Difficulty'], aggfunc=lambda x: ', '.join(x))
names_by_diff_table
```

```python
# get rid of dupes in df
df = df.drop_duplicates(subset='Trail Name')
df.head()
```

```python
df.shape
```

```python
final_table = pd.pivot_table(data=df, values='Trail Name', index=['Park Name', 'Difficulty'], aggfunc=lambda x: ', '.join(list(set(x))))
final_table
```

```python
final_table.to_excel('final_output.xlsx')
```

```python
# make table join-able
table_to_join = pd.pivot_table(data=df, values='Trail Name', columns='Difficulty', index='Park Name', aggfunc=(lambda x: ', '.join(list(set(x)))))
table_to_join
```

```python
table_to_join.to_excel('table_to_join.xlsx')
```
